```python
# ARTI308 Lab 6 - Full Solution

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# ------------------------------
# 2. Load Dataset
# ------------------------------
ecom = pd.read_csv('Ecommerce Customers')

print("First 5 rows:")
print(ecom.head())

print("\nDataset Info:")
print(ecom.info())

print("\nStatistical Summary:")
print(ecom.describe())

# ------------------------------
# 3. Feature Selection
# ------------------------------
X = ecom[['Avg. Session Length',
          'Time on App',
          'Time on Website',
          'Length of Membership']]

y = ecom['Yearly Amount Spent']

# ------------------------------
# 4. Train/Test Split
# ------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)

# ------------------------------
# 5. Train Linear Regression Model
# ------------------------------
lm = LinearRegression()
lm.fit(X_train, y_train)

# ------------------------------
# 6. Model Coefficients
# ------------------------------
coeff_df = pd.DataFrame(lm.coef_, X.columns,
                        columns=['Coefficient'])

print("\nModel Coefficients:")
print(coeff_df)

# ------------------------------
# 7. Predictions
# ------------------------------
predictions = lm.predict(X_test)

# Scatter Plot
plt.figure()
plt.scatter(y_test, predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()

# Residual Histogram
plt.figure()
sns.histplot(y_test - predictions, bins=50)
plt.title("Residual Distribution")
plt.show()

# ------------------------------
# 8. Model Evaluation
# ------------------------------
print("\nModel Evaluation:")
print("MAE:", metrics.mean_absolute_error(y_test, predictions))
print("MSE:", metrics.mean_squared_error(y_test, predictions))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, predictions)))

print("\nLab Completed Successfully ")
```

```python
# ARTI308 - Lab 6

# Linear Regression on E-commerce Customers

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Load Dataset
ecom = pd.read_csv('Ecommerce Customers')

# Data Exploration
print("Dataset Shape:", ecom.shape)
print("\nMissing Values:\n", ecom.isnull().sum())

# Feature Selection
X = ecom[['Avg. Session Length',
          'Time on App',
          'Time on Website',
          'Length of Membership']]

y = ecom['Yearly Amount Spent']

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)

# Train Model
model = LinearRegression()
model.fit(X_train, y_train)

# Coefficients
coefficients = pd.DataFrame(model.coef_, X.columns,
                            columns=['Coefficient'])
print("\nModel Coefficients:\n")
print(coefficients)

# Predictions
predictions = model.predict(X_test)

# Visualization
plt.figure()
plt.scatter(y_test, predictions)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted")
plt.show()

# Evaluation
print("\nModel Evaluation:")
print("MAE:", metrics.mean_absolute_error(y_test, predictions))
print("MSE:", metrics.mean_squared_error(y_test, predictions))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, predictions)))

print("\nLab Completed Successfully.")
```
